```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#generating x,y with some noise
x=np.linspace(0,50,50)
y=np.linspace(0,50,50)
x+=np.random.uniform(-5,5,50)
y+=np.random.uniform(-5,5,50)
n=len(x)
```

```python
#visualising
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training data')
plt.show()
```

```python
x_mean=x.mean()
y_mean=y.mean()
a1=(((x - x_mean) * (y - y_mean)).sum())/(((x - x_mean)**2).sum())
a0=y_mean - (a1 * x_mean)
```

```python
#visualising
plt.plot(x,a0+a1*x,'-r',label='Predicted')
plt.scatter(x,y,label='Actual')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
a1=a0=0 #initialise
epoch=100
a=0.001 #learning rate
for i in range(epoch):
    p=a0+a1*x
    da0=(-2/n)*(np.sum(y-p))
    da1=(-2/n)*(np.sum(x*(y-p)))
    a0=a0-a*da0
    a1=a1-a*da1
p=a0+a1*x
```

```python
#visualising
plt.plot(x,p,'-r',label='Predicted')
plt.scatter(x,y,label='Actual')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training data')
plt.legend()
plt.grid(True)
plt.show()
```
