```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data for two classes
np.random.seed(0)
```

```python
# Number of data points for each class
num_samples = 100
```

```python

# Generate random samples for both classes
class1_data = np.random.randn(num_samples, 2) + np.array([2, 2])
class2_data = np.random.randn(num_samples, 2) + np.array([5, 5])
print(f"class1{class1_data}\n\nclass 2 {class2_data}")
```

```python
# Combine the data and labels
X = np.concatenate((class1_data, class2_data), axis=0)
y = np.concatenate((np.zeros(num_samples), np.ones(num_samples)))
print(f"{X}\n\n{y}")
```

```python
# Initialize weights
num_features = X.shape[1] + 1
weights = np.zeros(num_features) #index 0 is the bias
print(f"{num_features}\n\n{weights}")
```

```python
# Set the learning rate and number of iterations
learning_rate = 0.07
num_iterations = 100000
```

```python
# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Initialize an empty list to store the loss values
loss_history = []

# Training the logistic regression model
for i in range(num_iterations):
    z = np.dot(X, weights[1:]) + weights[0]
    predictions = sigmoid(z)
    gradient = np.dot(X.T, (predictions - y)) / len(y)
    weights[1:] -= learning_rate * gradient
    weights[0] -= learning_rate * np.sum(predictions - y) / len(y)
    loss = -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

    # Append the loss value to the history list
    loss_history.append(loss)

    # Compute the binary cross-entropy loss
    loss = -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

    # Print the loss for monitoring
    if i % 10000 == 0:
        print(f"Iteration {i}: Loss = {loss}")
```

```python
# Plot the loss curve
plt.plot(range(num_iterations), loss_history)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```

```python
# Plot the decision boundary
x_boundary = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_boundary = -(weights[0] + weights[1] * x_boundary) / weights[2]
#plotting: w[0]+w[1]*x + w[2]*y=0

# Plot the data points
plt.scatter(class1_data[:, 0], class1_data[:, 1], label='Class 1', marker='o')
plt.scatter(class2_data[:, 0], class2_data[:, 1], label='Class 2', marker='x')

# Plot the decision boundary
plt.plot(x_boundary, y_boundary, label='Decision Boundary', color='red')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
